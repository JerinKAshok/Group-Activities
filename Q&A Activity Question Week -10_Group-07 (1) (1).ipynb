# 8. Check the outliers and handle outliers in performance score using Percentiles.


```python
plt.boxplot(data['age'])# checking 'age' column
plt.title('Age')
```

- No outliers in 'age'.

```python
plt.boxplot(data['performance_score'])# checking 'performance_score' column
plt.title('performance_score')

Q1=np.quantile(data['performance_score'],0.25)
Q3=np.quantile(data['performance_score'],0.75)
print('Q1,Q3=')
print(Q1,'\n',Q3)
IQR=Q3-Q1
print('IQR=',IQR)

lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
print('lower_limit=',lower_limit)
print('upper_limit=',upper_limit)

# finding outliers
outlier=[]
for a in data['performance_score']:
    if (a<lower_limit or a>upper_limit):
        outlier.append(a)
print('outlier=',outlier)

# finding index of outliers
index_of_outliers=data.loc[data['performance_score']<lower_limit]
print('index_of_outliers\n')
index_of_outliers

#removing outliers
data.drop(['James Authur'],inplace=True)
#displaying the data
data
```

```python
plt.boxplot(data['performance_score'])#plotting boxplot after removing outliers
plt.title('New Boxplot of performance_score')
```
